```python
import cv2
import numpy as np

rect_start = None
rect_end = None


# Function to create mask to extract ROI of area in frame of video
def create_rect_mask():
    global rect_start, rect_end, frame
    mask = np.zeros_like(frame)
    if rect_start is not None and rect_end is not None:
        cv2.rectangle(mask, rect_start, rect_end, (255, 255, 255), cv2.FILLED)
    masked_image = cv2.bitwise_and(frame, mask)
    return masked_image


# Function to clear the rectangle
def clear_rect():
    global rect_start, rect_end
    rect_start = None
    rect_end = None
    print("Deleted rectangle")


# Event handler for mouse clicks
def click_event(event, x, y, flags, param):
    global rect_start, rect_end

    if event == cv2.EVENT_LBUTTONDOWN:
        rect_start = (x, y)

    elif event == cv2.EVENT_RBUTTONDOWN:
        rect_end = (x, y)
        cv2.rectangle(frame, rect_start, rect_end, (0, 255, 0), 2)

   
        


cv2.namedWindow('Image')
cv2.setMouseCallback('Image', click_event)

# Read video
cap = cv2.VideoCapture('D:\\NTA\\video.mp4')

while True:
    ret, frame = cap.read()

    # Draw rectangle
    if rect_start is not None and rect_end is not None:
        cv2.rectangle(frame, rect_start, rect_end, (0, 255, 0), 2)

    cv2.imshow('Image', frame)

    # Handle key presses
    key = cv2.waitKey(1)

    # Close the application
    if key & 0xFF == ord('q'):
        break

    # Create mask to extract area
    elif key == ord('s'):
        mask = create_rect_mask()
        cv2.imwrite("D:\\NTA\\screen.png", mask)
        print("Saved image")

    # Clear the rectangle
    elif key == ord('c'):
        clear_rect()
        
   
cap.release()
cv2.destroyAllWindows() 
```

```python
import cv2
import numpy as np

circle_center = None
circle_radius = None


# Function to create mask to extract ROI of area in frame of video
def create_circle_mask():
    global circle_center, circle_radius, frame
    mask = np.zeros_like(frame )
    if circle_center is not None and circle_radius is not None:
        cv2.circle(mask, circle_center, circle_radius, (255, 255, 255), cv2.FILLED)
    masked_image = cv2.bitwise_and(frame, mask)
    return masked_image


# Function to clear the circle
def clear_circle():
    global circle_center, circle_radius
    circle_center = None
    circle_radius = None
    print("Deleted circle")


# Event handler for mouse clicks
def click_event(event, x, y, flags, param):
    global circle_center, circle_radius

    if event == cv2.EVENT_LBUTTONDOWN:
        circle_center = (x, y)

    elif event == cv2.EVENT_RBUTTONDOWN:
        circle_radius = int(np.sqrt((x - circle_center[0]) ** 2 + (y - circle_center[1]) ** 2))
        cv2.circle(frame, circle_center, circle_radius, (0, 255, 0), 2)


cv2.namedWindow('Image')
cv2.setMouseCallback('Image', click_event)

# Read video
cap = cv2.VideoCapture('D:\\NTA\\video.mp4')

while True:
    ret, frame = cap.read()

    # Draw circle
    if circle_center is not None and circle_radius is not None:
        cv2.circle(frame, circle_center, circle_radius, (0, 255, 0), 2)

    cv2.imshow('Image', frame)

    # Handle key presses
    key = cv2.waitKey(1)

    # Close the application
    if key & 0xFF == ord('q'):
        break

    # Create mask to extract area
    elif key == ord('s'):
        mask = create_circle_mask()
        cv2.imwrite("D:\\NTA\\screen.png", mask)
        print("Saved image")

    # Clear the circle
    elif key == ord('c'):
        clear_circle()

cap.release()
cv2.destroyAllWindows()
```
